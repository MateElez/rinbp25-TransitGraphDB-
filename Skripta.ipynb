```python
import json
from datetime import datetime
%cd /content/drive/MyDrive/RiNBP i AN projekt
# Uƒçitaj GeoJSON
with open('export.geojson') as f:
    geojson_data = json.load(f)

# Transformacija
output_data = []
for feature in geojson_data['features']:
    transformed = {
        'route_id': feature['properties'].get('ref'),
        'name': feature['properties'].get('name'),
        'operator': feature['properties'].get('operator'),
        'type': feature['properties'].get('route'),
        'from': feature['properties'].get('from'),
        'to': feature['properties'].get('to'),
        'geometry': feature['geometry'],
        'osm_data': {
            'id': feature['properties'].get('@id'),
            'version': feature['properties'].get('public_transport:version')
        },
        'timestamp': datetime.now().isoformat() + 'Z'
    }
    output_data.append(transformed)

# Snimi transformirane podatke
with open('transformed_routes.json', 'w') as f:
    json.dump(output_data, f, indent=2)

# Download fajla
from google.colab import files
files.download('transformed_routes.json')
```
